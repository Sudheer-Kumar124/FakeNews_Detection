```python
import pandas as pd
import nltk
from nltk.corpus import stopwords



# Example combining 'title' and 'description' into 'text' column (replace with your own combining logic)
df['text'] = df['title'].astype(str) + ' ' + df['description'].astype(str)

# Download NLTK stopwords list
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Example cleaning function to remove stopwords
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Apply the cleaning function to the 'text' column
df['text'] = df['text'].apply(remove_stopwords)

# Print the cleaned text data
print(df['text'].head())
```

```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize



# Example combining 'title' and 'description' into 'text' column (replace with your own combining logic)
df['text'] = df['title'].astype(str) + ' ' + df['description'].astype(str)

# Download NLTK resources (if needed)
nltk.download('punkt')

# Initialize PorterStemmer
porter = PorterStemmer()

# Example stemming function
def stem_text(text):
    words = word_tokenize(text)
    stemmed_words = [porter.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Apply the stemming function to the 'text' column
df['text'] = df['text'].apply(stem_text)

# Print the stemmed text data
print(df['text'].head())
```

```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize



# Example combining 'title' and 'description' into 'text' column (replace with your own combining logic)
df['text'] = df['title'].astype(str) + ' ' + df['description'].astype(str)

# Download NLTK resources (if needed)
nltk.download('punkt')
nltk.download('wordnet')

# Initialize WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Example lemmatization function
def lemmatize_text(text):
    words = word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

# Apply the lemmatization function to the 'text' column
df['text'] = df['text'].apply(lemmatize_text)

# Print the lemmatized text data
print(df['text'].head())
```
