```python
import re
```

```python
def wordopt(text):
    #convert into lowercase
    text = text.lower()
    
    #remove urls 
    text = re.sub(r'https?://|S+|www\.\S+','',text)
    
    #remove html tags
    text = re.sub(r'<.*?>','',text)
    
    #remove punctuation
    text = re.sub(r'[^\w\s]','',text)
    
    #remove digits
    text = re.sub(r'\d','',text)
    
    #remove newline characters
    text = re.sub(r'\n',' ',text)
    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(xv_train,y_train)
```

```python
pred_lr = lr.predict(xv_test)
lr.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
pred_dtc = dtc.predict(xv_test)
```

```python
dtc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
def output_label(n):
    if n==0:
        return "it is a fake news"
    elif n == 1:
        return "it is a genuine news"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]} #corrected syntax for defining dictionary
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test) #assuming 'vectorization' is your vectorizer object
    pred_lr = lr.predict(new_xv_test)
    res = ""
    if(pred_lr[0]==0):
         res = res + " LR Prediction : it is a fake news\n"
    elif pred_lr[0] == 1:
        res=res+" LR Prediction : it is a genuine news\n"
    pred_gbc = gbc.predict(new_xv_test)
    if(pred_gbc[0]==0):
         res = res + " GBC Prediction : it is a fake news\n"
    elif pred_gbc[0] == 1:
        res=res+" GBC Prediction : it is a genuine news\n"
    pred_rfc = rfc.predict(new_xv_test)
    if(pred_rfc[0]==0):
         res = res + " RFC Prediction : it is a fake news"
    elif pred_rfc[0] == 1:
        res=res+" RFC Prediction : it is a genuine news"
    #str = "LR prediction :",output_label(pred_lr[0]), " GBC Prediction : ",output_label(pred_gbc[0])," RFC Prediction : ",output_label(pred_rfc[0])
    #return "\n\nLR Prediction : {}   \nGBC Prediction : {}  \nRFC Prediction : {}".format(output_label(pred_lr[0]),output_label(pred_gbc[0]),output_label(pred_rfc[0]))
    # str = "LR prediction : ",output_label(pred_lr[0])
    # return str
    return res
```

```python
import tkinter as tk
from tkinter import ttk

def process_input():
    # Get the user input from the input field
    user_input = input_entry.get()
    
    # Process the input (e.g., transform the input or calculate a result)
    # Here, as an example, we'll just convert the input to uppercase
    result = manual_testing(user_input)
    
    # Update the output label with the processed result
    output_label.config(text=f"Output:\n {result}")

def on_close():
    # Handle the close event
    window.destroy()

# Create the main window
window = tk.Tk()
window.title("Fake News Detection Application")

# Set the style
style = ttk.Style()
style.theme_use("clam")

# Create a frame to hold the input and button
frame = ttk.Frame(window)
frame.pack(pady=20, padx=20)

# Create an input label and entry field
input_label = ttk.Label(frame, text="Enter article:", font=("Arial", 12))
input_label.pack(side=tk.LEFT, padx=5)

input_entry = ttk.Entry(frame, font=("Arial", 12))
input_entry.pack(side=tk.LEFT, padx=5)

# Create a button to trigger the input processing
process_button = ttk.Button(frame, text="Process", command=process_input)
process_button.pack(side=tk.LEFT, padx=5)

# Create an output label to display the result
output_label = ttk.Label(window, text="Output: ", font=("Arial", 12))
output_label.pack(pady=20, padx=20)

# Set up the close button event
window.protocol("WM_DELETE_WINDOW", on_close)

# Run the main loop
window.mainloop()
```
